## Where did that 29 come from?

The prime factors for 91 were 7, 13.

Let's take a modulus just a smidge smaller.... the product of 7-1 & 13-1

```python
(7-1) * (13-1)
```

Then, let's take a prime number that doesn't divide into 72.

```python
def check_if_prime(num):
    roots = []
    all_numbers_lower = list(range(1,num+1))
    all_numbers_lower.reverse()
    for denominator in all_numbers_lower:
        if num/denominator == int(num/denominator):
            roots.append(denominator)
    
    if len(roots) == 2:
        return num,True
    else:
        return num,False
    

def all_primes_up_to(num):
    
    primes = []
    for i in range(1,num+1):
        if check_if_prime(i)[1] == True:
            primes.append(i)

    return primes    


for prime_we_need in all_primes_up_to(72):
    if 72 % prime_we_need != 0:
        break

prime_we_need
```

That's our encription value.

So how do we find a value of `d` now we have `prime_we_need`, such that `d` multiplied by `prime_we_need` is a value of `1` on our new `72` modulus.

```python
n = 1
while n < 72:
    if ((72 * n)+1) % 5 == 0:
        break
    n += 1

    
d = (72 * n + 1) / 5

d
```

# Bingo! That's 29. 

So why does that work?

If we subtract 1 from $d \times 5$, we get `144`. `144` is also a multiple of `6` & `12`, & `72`.

This bit now means we can implement Fermat's Little Theorem.

Let's call `prime_we_need` `e`.

$ (e \times d) - 1 = m( (p-1) (q-1) )$ with p being 7 and q being 13

$ \frac{(e \times d) - 1}{m} = (p-1) (q-1)$

```python
((prime_we_need * d) - 1) % ((7-1) * (13-1)) == 0
```

## Then we come to Fermat's little theorem. 

If $f$ is a prime and $a$ is any integer not divisible by $f$, then $a^{f  1}  1$ is divisible by $f.$


### So $m^{ed-1} \div (1)(1)  \equiv m$

```python
m = ((prime_we_need * d) - 1) / ((7-1) * (13-1))
```

```python
def fermat_test(a, p):
    """Tests if a^p is congruent to a modulo p using Fermat's Little Theorem."""
    if not check_if_prime(p)[1]:
        raise ValueError("p must be a prime number")
    return pow(a, p, p) == a
```
