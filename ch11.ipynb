### Anyway, what you have to remember in terms of imaginary numbers is that they are basically $\sqrt -1 = i$ with value of `i` being the negative of the negative square root. So $\sqrt-9 = 3i$ because the negative of `-9` square rooted is 3....

## Anyway, imaginary numbers are incredibly important in their role on mathematics. You literally wouldn't be able to read these very words without them. 

### One of the cool things about them is that they square easy. For example:

# $(-4 + 7i)^2 = (-4 + 7i) \times (-4 + 7i) = (-33 - 56i)$

## Imaginary numbers are also important when it comes to defining the most beautiful function in mathematics.

# $e^{\pi i} = -1$

## Might not be immediately obvious, but the reason why this is such a beautifully function is that it draws it so much different mathematics, so simply. It relates $\pi$ , which is meant to be about circles, to Euler's number, which is meant to be about natural logarithms, to a perfectly round minus 1. All through the power of imaginary numbers.
