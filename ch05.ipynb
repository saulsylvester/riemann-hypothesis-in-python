# Back to Ancient Greece...

So Eratosthenes of Cyrene (which is actually in Libya) was one of the librarians at the Great Library of Alexandria. Around 70 years after our hero Euclid, Eratosthenes devised a method (quite like our `check_if_prime` from chapter 2) to determine whether a number is prime. 

So for example, if we wanted to the first 120 numbers we can to the same.

```python
first_120_numbers = list(range(2,120+1))

first_120_numbers
```

OK, then we know primes aren't divisible by anything other than itself and 1. So if a value is divisible by 2, then it's not a prime (unless it's 2 itself of course).

```python
denominator = 2
new_list = [num for num in first_120_numbers if num/denominator != int(num/denominator) or num == denominator]

new_list
```

OK, then from this new list, we can remove those that are divisible by 3...



```python
denominator = 3
new_list_removing_multiples_of_3 = [num for num in new_list if num/denominator != int(num/denominator) or num in [2,3]]

new_list_removing_multiples_of_3
```

### Now, `new_list_removing_multiples_of_3` accurately reflects the number of primes up 3 and the next number which is 5, we could do the same for all the numbers remaining in the list each time. So after 5, tackle 7, then 11, all the way up until 23 where the next number `25` wouldn't be in the list. That's basically how the genius Eratosthenes did it.

OK, so let's bring this logic to the Riemann Zeta function. 

### If I divided both sides by $2^s$ (or multiplied them both by $\frac{1}{2^s}$): 

$ \zeta(s) = 1+\frac{1}{2^s}+\frac{1}{3^s}+\frac{1}{4^s}+\frac{1}{5^s}+\frac{1}{6^s}+\frac{1}{7^s}+\frac{1}{8^s}+\frac{1}{9^s}+\frac{1}{10^s}+\frac{1}{11^s}...$
 
would turn to:
 
$ \frac{1}{2^s}\zeta(s) = \frac{1}{2^s}+\frac{1}{4^s}+\frac{1}{6^s}+\frac{1}{8^s}+\frac{1}{10^s}+\frac{1}{12^s}+\frac{1}{14^s}+\frac{1}{16^s}+\frac{1}{18^s}+\frac{1}{20^s}...$
 
 
If that doesn't make sense, let me make it a little smaller:

$ \zeta(s) = 1+\frac{1}{2^s}+\frac{1}{3^s}+\frac{1}{4^s}...$ -> $ \frac{1}{2^s}\zeta(s) = \frac{1}{2^s}+\frac{1}{4^s}+\frac{1}{6^s}+\frac{1}{8^s}$


## Now if you remember the rule of $ (x \times y)^n = x^n \times y^n $ we can do something a little tricky here. We can take this new expression of $ \frac{1}{2^s}\zeta(s)$ away from the original $ \zeta(s)$ series. 

$ \zeta(s) - \frac{1}{2^s}\zeta(s) = (1 - \frac{1}{2^s})\zeta(s)$


$(1 - \frac{1}{2^s})\zeta(s) = 1+\frac{1}{3^s}+\frac{1}{5^s}+\frac{1}{7^s}+\frac{1}{9^s}+\frac{1}{11^s}+\frac{1}{13^s}+\frac{1}{15^s}+\frac{1}{17^s}+\frac{1}{19^s}+\frac{1}{21^s}...$


And by removing all the series of the second $\frac{1}{2^s}\zeta(s)$ sequence, we now have a sequence of just odd denominators.... I'm sure you see where we're going with this...

Multipliying both sides of this new sequence by $\frac{1}{3^s}\zeta(s)$, we get...


$\frac{1}{3^s}(1 - \frac{1}{2^s})\zeta(s) = 1+\frac{1}{3^s}+\frac{1}{9^s}+\frac{1}{15^s}+\frac{1}{21^s}+\frac{1}{27^s}+\frac{1}{33^s}+\frac{1}{39^s}+\frac{1}{45^s}+\frac{1}{51^s}+\frac{1}{57^s}...$


Which we can also subtract from the second series:

$(1 - \frac{1}{3^s})(1 - \frac{1}{2^s})\zeta(s) = 1+\frac{1}{5^s}+\frac{1}{7^s}+\frac{1}{11^s}+\frac{1}{13^s}+\frac{1}{17^s}+\frac{1}{19^s}+\frac{1}{23^s}+\frac{1}{25^s}+\frac{1}{29^s}+\frac{1}{31^s}...$


We can actually do this for all the primes, similar to Eratosthenes' sieve.


### 5? 
$(1 - \frac{1}{5^s})(1 - \frac{1}{3^s})(1 - \frac{1}{2^s})\zeta(s) = 1+\frac{1}{7^s}+\frac{1}{11^s}+\frac{1}{13^s}+\frac{1}{17^s}+\frac{1}{19^s}+\frac{1}{23^s}+\frac{1}{29^s}+\frac{1}{31^s}...$


### And then 7?

$(1 - \frac{1}{7^s})(1 - \frac{1}{5^s})(1 - \frac{1}{3^s})(1 - \frac{1}{2^s})\zeta(s) = 1+\frac{1}{11^s}+\frac{1}{13^s}+\frac{1}{17^s}+\frac{1}{19^s}+\frac{1}{23^s}+\frac{1}{29^s}+\frac{1}{31^s}...$


### What about all the way to 997?

$(1 - \frac{1}{997^s})(1 - \frac{1}{991^s})...(1 - \frac{1}{5^s})(1 - \frac{1}{3^s})(1 - \frac{1}{2^s})\zeta(s) = 1+\frac{1}{1009^s}+\frac{1}{1013^s}+\frac{1}{1019^s}+\frac{1}{1021^s}$

```python
def check_if_prime(num):
    roots = []
    all_numbers_lower = list(range(1,num+1))
    all_numbers_lower.reverse()
    for denominator in all_numbers_lower:
        if num/denominator == int(num/denominator):
            roots.append(denominator)
    
    if len(roots) == 2:
        return num,True
    else:
        return num,False
    
check_if_prime(1009)
```

# Hopefully you can see where this is going. In theory if we were able to put every prime number throught this, we'd end up with something like:

$...(1 - \frac{1}{17^s})(1 - \frac{1}{13^s})(1 - \frac{1}{11^s})(1 - \frac{1}{7^s})(1 - \frac{1}{5^s})(1 - \frac{1}{3^s})(1 - \frac{1}{2^s})\zeta(s) = 1$

### Imagine can also divide both sides by $(1 - \frac{1}{2^s})$

$...(1 - \frac{1}{17^s})(1 - \frac{1}{13^s})(1 - \frac{1}{11^s})(1 - \frac{1}{7^s})(1 - \frac{1}{5^s})(1 - \frac{1}{3^s})\zeta(s) = 1\times\frac{1}{1 - \frac{1}{2^s}}$

### And by $1 - \frac{1}{3^s}$

$...(1 - \frac{1}{17^s})(1 - \frac{1}{13^s})(1 - \frac{1}{11^s})(1 - \frac{1}{7^s})(1 - \frac{1}{5^s})\zeta(s) = 1\times\frac{1}{1 - \frac{1}{2^s}}\times\frac{1}{1 - \frac{1}{3^s}}$

### In fact,  if we follow some of the indice flipping rules we looked at in the last chapter, we can do them all and drop the first 1:

$\zeta(s) = \frac{1}{1 - \frac{1}{2^s}}\times\frac{1}{1 - \frac{1}{3^s}}\times\frac{1}{1 - \frac{1}{5^s}}\times\frac{1}{1 - \frac{1}{7^s}}\times\frac{1}{1 - \frac{1}{11^s}}...$


### And also:

$\zeta(s) = (1-2^s)^{-1}(1-3^s)^{-1}(1-5^s)^{-1}(1-7^s)^{-1}(1-11^s)^{-1}...$


### This is essentially a great function (also known as the Euler product ), which can be shortened to:

$\zeta(s) = \Pi_{p}(1-p^s)^{-1}$

### Which comes to mean that:
### the Zeta of s equals the product over all primes of one minus p to the minus s, to the minus one.

## And therefore if we can remember the last chapter

$\sum n^{-s} = \Pi_{p}(1-p^s)^{-1}$

The part that is probably most fantastic about all this is the fact that on the left-hand side of we have an infinite sum running through all the positive whole numbers 1, 2, 3, 4, 5, 6, â€¦, while on the right-hand side we have an infinite product running through all the prime numbers 2, 3, 5, 7, 11, 13.
