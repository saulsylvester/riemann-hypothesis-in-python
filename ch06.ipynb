```python
result = 1.0
d = 3
plus = False
counter = 0
# while round(sin(result * 4),16) != 0 or 
# & result < (pi+(1/10**16)):
# while counter < 20:

while counter < 600:
    if plus == True:
        print(f"{result}+ 1/{d} => {(result+ 1/d) * 4}")
        result = result + 1/d
        
    else:
        print(f"{result}- 1/{d} => {(result- 1/d) * 4}")
        result = result - 1/d
        
    plus = not plus
    d += 2
    counter += 1
```
